```python
x = np.linspace(1,51,51)
lamb = 12
```

```python
def poissonPDF_adj(lamb, k,theta):
    return np.exp(-lamb) *((lamb**k) / fac(int(k)))**theta
```

```python
([poissonPDF_adj(lamb, i, 1) for i in x])
```

```python
y_08 = [poissonPDF_adj(lamb, i, 0.8)/np.sum([poissonPDF_adj(lamb, i, 0.8) for i in x]) for i in x]
y_1 = [poissonPDF_adj(lamb, i, 1)/np.sum([poissonPDF_adj(lamb, i, 1) for i in x]) for i in x]
y_12 = [poissonPDF_adj(lamb, i, 1.2)/np.sum([poissonPDF_adj(lamb, i, 1.2) for i in x]) for i in x]
```

```python
plt.plot(y_08,label='0.8')
plt.plot(y_1,label='1') 
plt.plot(y_12,label='1.2')
plt.legend()
```

```python
for i in [y_08, y_1, y_12]:
    vals = []
    for a,b in zip(x, i):
        for _ in range(int(b*1000000)):
            vals.append(a)
    print(len(vals))    
    print(np.mean(vals),np.var(vals),np.var(vals)/np.mean(vals))
        
```
